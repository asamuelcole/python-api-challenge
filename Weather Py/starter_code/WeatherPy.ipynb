```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set up API url and lists to store information
url = "http://api.openweathermap.org/data/2.5/weather?"
City_ID = 1
Set_ID = 1
name_list = []
lon_list = []
lat_list = []
max_temp_list = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
country_list = []
date_list = []

# API details
for city in cities:
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=imperial"
    
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    
# store information into lists and print city records
    try:
        name_list.append(weather_json['name'])
        lon_list.append(weather_json['coord']['lon'])
        lat_list.append(weather_json['coord']['lat'])
        max_temp_list.append(weather_json['main']['temp_max'])
        humidity_list.append(weather_json['main']['humidity'])
        cloudiness_list.append(weather_json['clouds']['all'])
        wind_speed_list.append(weather_json['wind']['speed'])
        country_list.append(weather_json['sys']['country'])
        date_list.append(weather_json['dt'])
        print(f"Processing record number {City_ID} of set {Set_ID}| {city}")
        print("-----------------------------------------------------------")
        City_ID = City_ID + 1
        if City_ID > 49:
            City_ID = 1
            Set_ID = Set_ID + 1
    except KeyError:
        print(f"City not found. Skipping {city}.")
        print("---------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# combine lists into DataFrame
cities_data = {
    "City": name_list,
    "Cloudiness (%)": cloudiness_list,
    "Country": country_list,
    "Date": date_list,
    "Humidity (%)": humidity_list,
    "Latitude": lat_list,
    "Longitude": lon_list,
    "Max Temp (F)": max_temp_list,
    "Wind Speed (mph)": wind_speed_list,
}

cities_df = pd.DataFrame(cities_data)
cities_df.index.name = 'City ID'
cities_df.head()
```

```python
# export data into csv file
cities_df.to_csv("../output_data/cities2.csv", index=True, header=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Max Temp vs. Latitude with linear regression
x_values = northern_df['Latitude']
y_values = northern_df['Max Temp (F)']
nor_lat_temp = northern_df.plot(kind='scatter', x='Latitude', y='Max Temp (F)')
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.savefig("../output_data/nor_lat_temp")

```
